Importing libraries

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler, LabelEncoder, OneHotEncoder
from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
```

```python
dataset = pd.read_csv('Breast_Cancer.csv')
```

```python
dataset_nan = dataset.isnull().values.any()
dataset_nan
```

```python
dataset.columns
```

Dataset have no NaN values

LabelEncoder

```python
label_encoder = LabelEncoder()
le_columns=['T Stage ','N Stage','6th Stage','Status']

dataset[le_columns] = dataset[le_columns].apply(LabelEncoder().fit_transform)
```

OneHotEncoder

```python
oh_encoder=OneHotEncoder()
encoded_data = oh_encoder.fit_transform(dataset[['Race', 'Marital Status', 'differentiate','A Stage','Estrogen Status','Progesterone Status']])
encoded_array = encoded_data.toarray()
encoded_df = pd.DataFrame(encoded_array, columns=oh_encoder.get_feature_names_out(['Race', 'Marital Status', 'differentiate','A Stage','Estrogen Status','Progesterone Status']))
encoded_df
```
