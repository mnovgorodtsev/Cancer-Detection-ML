Importing libraries

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
```

```python
dataset = pd.read_csv('Breast_Cancer.csv')
```

```python
dataset_nan = dataset.isnull().values.any()
dataset_nan
```

```python
dataset.columns
```

Dataset have no NaN values

LabelEncoder

```python
label_encoder = LabelEncoder()
le_columns=['T Stage ','N Stage','6th Stage','Status']

dataset[le_columns] = dataset[le_columns].apply(LabelEncoder().fit_transform)
```

```python
dataset
```

OneHotEncoder

```python
oh_encoder=OneHotEncoder()
encoded_data = oh_encoder.fit_transform(dataset[['Race', 'Marital Status', 'differentiate','A Stage','Estrogen Status','Progesterone Status']])
encoded_array = encoded_data.toarray()
encoded_df = pd.DataFrame(encoded_array, columns=oh_encoder.get_feature_names_out(['Race', 'Marital Status', 'differentiate','A Stage','Estrogen Status','Progesterone Status']))
encoded_df
```

Y and column which were encoded by OneHotEncoder drop

```python
y = dataset.iloc[:, -1]
dataset = dataset.drop(columns=['Race', 'Marital Status', 'differentiate','A Stage','Estrogen Status','Progesterone Status','Status'])
```

Merging two dataframes

```python
dataset = pd.concat([dataset, encoded_df, y], axis=1)
dataset['Status'] = (~dataset.Status.astype(bool)).astype(int)
```
